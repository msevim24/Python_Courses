```python
one = pd.DataFrame({
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5'],
   'Marks_scored':[98,90,87,69,78]},
   index=[1,2,3,4,5])
```

```python
two = pd.DataFrame({
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5'],
   'Marks_scored':[89,80,79,97,88]},
   index=[1,2,3,4,5])
```

```python
pd.concat([one, two])
```

```python
pd.concat([one, two], ignore_index=True)
```

```python
df2=pd.concat([one, two], keys=["x", "y"])
```

```python
df2.loc["x"]
```

```python
df2.loc["y"]
```

```python
pd.concat([one, two], axis=1)
```

```python
one.append([two, one, two])
```

```python
one.append([two, one, two], ignore_index=True)
```

```python
x = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                     'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']})
   
y = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']}) 
```

```python
x
```

```python
y
```

```python
pd.merge(x, y, how="inner", on="key")
```

```python
pd.merge(x, y, how="outer", on="key")
```

```python
pd.merge(x, y, how = 'left', on = 'key')
```

```python
pd.merge(x, y, how = 'right', on = 'key')
```

```python
df11 = pd.DataFrame({'lkey': ['x', 'y', 'z', 'x'],
                    'lvalue': [2, 3, 5, 7]})
df12 = pd.DataFrame({'rkey': ['a', 'x', 'z', 'b'],
                    'rvalue': [7, 8, 9, 10]})
```

```python
df11
```

```python
df12
```

```python
pd.merge(df11, df12, left_on="lkey", right_on="rkey", how="outer")
```

```python
pd.merge(df11, df12, left_on = "lkey", right_on = "rkey", how = "left")
```

```python
pd.merge(df11, df12, left_on = "lkey", right_on = "rkey", how = "right")
```

```python
import pandas as pd
import numpy as np
 
N=20
df = pd.DataFrame({
   'A': pd.date_range(start='2016-01-01',periods=N,freq='D'),
   'x': np.linspace(0,stop=N-1,num=N),
   'y': np.random.rand(N),
   'C': np.random.choice(['%%Low','((Medium','^High'],N).tolist(),
   'D': np.random.normal(100, 10, size=(N)).tolist()
   })
```

```python
df
```

```python
for i in range(len(df.C.values)):
    df.C.values[i]=df.C.values[i].strip("%^(")
```
